```python
df  = pd.read_csv('data_final_11.csv' , sep=',' )
```

```python
 df['logic'] = np.where(df['diagnosis'] == 'M',0.2,0);
```

```python
condition1 = (df['logic']== 0.2)
condition2 = (df['logic']== 0)
df2 = df[condition1]
df3 = df[condition2]
```

```python
def plot_binary_M_B(a):
    print('INTERVAL OF AMBIGUITY -> (',df2[a].min(),',',df3[a].max(),')')
    df.plot.scatter(x=a, y='logic');
    df[a].plot.kde(color='red')
    print("probability that radius we get will of M type :",(len(df2))/(len(df2)+len(df3)))
```

```python
def plot_binary_M_B1(a):
    print('INTERVAL OF AMBIGUITY -> (',df2[a].min(),',',df3[a].max(),')')
    df.plot.scatter(x=a, y='logic');
    print("probability that radius we get will of M type :",(len(df2))/(len(df2)+len(df3)))
```

# Observation from Radius mean plot :
## 1. There is a high chance that a person is melingin if mean radius is more than 17.85
## 2. There is a low chance that a person is  benign if mean radius is less than 10.38
## 3. Result can't be concluded if mean radius lies in the ambiguity region.

```python
plot_binary_M_B('radius_mean')
```
